# Metacell process using SEACells
SEACells must be installed to run the code below. The installation is available in https://github.com/dpeerlab/SEACells

```python
import os
import pandas as pd
import numpy as np

import scanpy as sc
import pyranges as pr
import warnings

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import data generated from ArchR

```python
# Read binning data which is named TileMatrix data
from scipy.io import mmread
TileMatrix = mmread('./metacell_input_data/TileMatrix.mtx')
```

```python
# Show binning object
TileMatrix
```

```python
# Get dimensions of Bin x Cell Matrix
dimensions = TileMatrix.shape
print(dimensions)
```

```python
# Read Cell information
cells = pd.read_csv('./metacell_input_data/cells.csv', index_col=0).iloc[:, 0]
```

```python
# Read Bin information
bins = pd.read_csv('./metacell_input_data/tileMatrix_rowData.csv', index_col=0).iloc[:, :]
```

```python
bins
```

```python
# Calculate the end index
tileSize = 500000
bins['end'] = bins['start'] + tileSize - 1

# Convert to integer
bins['start'] = bins['start'].astype(int)
bins['end'] = bins['end'].astype(int)

# Create the desired index
bins.index = (
    bins['seqnames'] + '_' +
    bins['idx'].astype(str) + ':' +
    bins['start'].astype(str) + '-' +
    bins['end'].astype(str)
)

# Display the DataFrame
print(bins.head())
```

```python
# Create AnnData
ad = sc.AnnData(TileMatrix.T)
ad.obs_names = cells
ad.var_names = bins.index
```

```python
# Show AnnData Object
ad
```

```python
# For AnnData, the data is stored in .X attribute.
# converts the .X attribute of an AnnData object ad to a Compressed Sparse Row (CSR) matrix format.
ad.X = ad.X.tocsr()
```

```python
# Add X_svd for the MetaCell process.
# X_svd means singular value decomposition (SVD) of the expression matrix X
ad.obsm['X_svd'] = pd.read_csv('./metacell_input_data/svd.csv', index_col=0).loc[ad.obs_names, : ].values
```

```python
cell_meta = pd.read_csv('./metacell_input_data/cell_metadata.csv', index_col=0).loc[ad.obs_names, : ]
for col in cell_meta.columns:
    ad.obs[col] = cell_meta[col].values
```

```python
# Show ad object after adding the objects and observements.
ad
```

```python
# Leiden and UMAP to get the Cell Type (Clusters)
warnings.filterwarnings('ignore')
sc.pp.neighbors(ad, use_rep='X_svd')
sc.tl.umap(ad)
sc.tl.leiden(ad)
warnings.filterwarnings('default')
```

```python
sc.pl.scatter(ad, basis='umap', color=['leiden', 'Clusters'])
```

```python
# Subset the AnnData object
# Get the data for Cell Type C3 and Cell Type C4
ad_C3 = ad[ad.obs['Clusters'] == 'C3', :]
ad_C4 = ad[ad.obs['Clusters'] == 'C4', :]

# Verify the result
print(ad_C3)
print(ad_C4)

# Convert to dense matrix
dense_matrix_C3 = ad_C3.X.todense()
# Convert to DataFrame for better visualization
df_C3 = pd.DataFrame(dense_matrix_C3)
print(df_C3)
# Convert DataFrame to NumPy array
array_C3 = df_C3.to_numpy()
# Save as .npy file
# feature_C3.pny is the Cell x Bin Matrix of Cell Type C3
np.save('./feature/feature_C3.npy', array_C3)

dense_matrix_C4 = ad_C4.X.todense()
# Convert to DataFrame for better visualization
df_C4 = pd.DataFrame(dense_matrix_C4)
print(df_C4)

# Convert DataFrame to NumPy array
array_C4 = df_C4.to_numpy()
# Save as .npy file
# feature_C4.pny is the Cell x Bin Matrix of Cell Type C4
np.save('./feature/feature_C4.npy', array_C4)
```

# Start Computation Metacell
Need to install SEACells firstly.
The installation is available in https://github.com/dpeerlab/SEACells

```python
# Calculate the Metacell for Cell Type C3
```

```python
# Need to install SEACells firstly.
# The installation is available in https://github.com/dpeerlab/SEACells
import SEACells
```

```python
## User defined parameters

## Core parameters to define the Final MetaCell numbers. 
# The recommendation is average 75 cells in one Metacell. Because there are 1351 cells in C3, the final Metalcell number will be 18.
n_SEACells_C3 = 18
build_kernel_on = 'X_svd' # key in ad.obsm to use for computing metacells
                          # This would be replaced by 'X_svd' for ATAC data

## Additional parameters
n_waypoint_eigs = 10 # Number of eigenvalues to consider when initializing metacells


## Core parameters
# The recommendation is average 75 cells in one Metacell. Because there are 1272 cells in C4, the final Metalcell number will be 16.
n_SEACells_C4 = 16
build_kernel_on = 'X_svd' # key in ad.obsm to use for computing metacells
                          # This would be replaced by 'X_svd' for ATAC data
```

```python
# To build the model of Cell Type C3 and Cell Type C4.
model_C3 = SEACells.core.SEACells(ad_C3,
                  build_kernel_on=build_kernel_on,
                  n_SEACells=n_SEACells_C3,
                  n_waypoint_eigs=n_waypoint_eigs,
                  convergence_epsilon = 1e-5)

model_C4 = SEACells.core.SEACells(ad_C4,
                  build_kernel_on=build_kernel_on,
                  n_SEACells=n_SEACells_C4,
                  n_waypoint_eigs=n_waypoint_eigs,
                  convergence_epsilon = 1e-5)
```

```python
# Create kenerl matrix for metacell.
model_C3.construct_kernel_matrix()
M_C3 = model_C3.kernel_matrix

model_C4.construct_kernel_matrix()
M_C4 = model_C4.kernel_matrix
```

```python
# Initialize archetypes
model_C3.initialize_archetypes()

# Initialize archetypes
model_C4.initialize_archetypes()
```

```python
# Plot the initilization to ensure they are spread across phenotypic space
SEACells.plot.plot_initialization(ad_C3, model_C3)


# Plot the initilization to ensure they are spread across phenotypic space
SEACells.plot.plot_initialization(ad_C4, model_C4)
```

```python
# Start interation and get convergence.
# For C3
model_C3.fit(min_iter=10, max_iter=50)
print('Complete C3 interation\n')
# For C4
model_C4.fit(min_iter=10, max_iter=50)
print('Complete C4 interation\n')
```

```python
# Check for convergence for C3 Metacell
model_C3.plot_convergence()

# Check for convergence for C4 Metacell
model_C4.plot_convergence()
```

```python
# Check Metacell head data.
ad_C3.obs[['SEACell']].head()

ad_C4.obs[['SEACell']].head()
```

```python
# Use hard assignment for metacell
model_C3.get_hard_assignments().head()
model_C4.get_hard_assignments().head()
```

```python
# Get Metacell data
SEACell_ad_C3 = SEACells.core.summarize_by_SEACell(ad_C3, SEACells_label='SEACell', summarize_layer='X')
SEACell_ad_C3

SEACell_ad_C4 = SEACells.core.summarize_by_SEACell(ad_C4, SEACells_label='SEACell', summarize_layer='X')
SEACell_ad_C4
```

```python
# Show metacell object in .X layer
SEACell_ad_C3.X

SEACell_ad_C4.X
```

```python
import pandas as pd

# Convert to dense format (if small enough)
dense_matrix_C3 = SEACell_ad_C3.X.todense()
# Convert to DataFrame for better visualization
df_C3 = pd.DataFrame(dense_matrix_C3)

# Display the first few rows
df_C3.head()


# Convert to dense format (if small enough)
dense_matrix_C4 = SEACell_ad_C4.X.todense()
# Convert to DataFrame for better visualization
df_C4 = pd.DataFrame(dense_matrix_C4)

# Display the first few rows
df_C3.head()
df_C4.head()
```
